```python
!wget -O retail.csv https://people.sc.fsu.edu/~jburkardt/data/csv/hw_200.csv
```

```python
df = spark.read.csv("retail.csv", header=True, inferSchema=True)
```

```python
df.printSchema()
```

```python
df.show(5)
```

```python
df.count()
```

```python
df.describe().show()
```

```python
from pyspark.sql.functions import col

df.select([col(c) for c in df.columns]).show(5)
```

```python
spark
```

```python
df.printSchema()
```

```python
df.show(10, truncate=False)
```

```python
df.count()
```

```python
from pyspark.sql.functions import col, sum

df.select([
    sum(col(c).isNull().cast("int")).alias(c)
    for c in df.columns
]).show()
```

```python
df.describe().show()
```

```python
from pyspark.sql.functions import monotonically_increasing_id

df = df.withColumn("row_id", monotonically_increasing_id())
df.show(5)
```

```python
df.createOrReplaceTempView("retail_data")
```

```python
spark.sql("""
SELECT *
FROM retail_data
LIMIT 5
""").show()
```

```python
spark.sql("""
SELECT COUNT(*) AS total_records
FROM retail_data
""").show()
```

```python
df_repartitioned = df.repartition(4)
df_repartitioned.rdd.getNumPartitions()
```
